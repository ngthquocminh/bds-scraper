```python
from pprint import pprint

def exist1(url_hash):
    doc = db.database.final_container.query_items(
        query='SELECT * FROM c where c.url_hash="%s"'%(url_hash),
        enable_cross_partition_query=True)
    return len(list(doc)) > 0

def exist2(url_hash):
    doc = db.database.container.query_items(
        query='SELECT * FROM c where c.url_hash="%s"'%(url_hash),
        enable_cross_partition_query=True)
    return len(list(doc)) > 0

def save(item):
    return db.database.container.upsert_item(item)
    
def update(item):
    return db.database.update_container.upsert_item(item)
```
